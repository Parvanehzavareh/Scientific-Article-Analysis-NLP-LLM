# Text Preprocessing

## Function to Apply Lemmatization and Stemming

```python
def lemmatize_stemming(text):
    return WordNetLemmatizer().lemmatize(text, pos='v')


def preprocess(text):
    result = []
    for token in gensim.utils.simple_preprocess(text):
        if token not in gensim.parsing.preprocessing.STOPWORDS and len(token) > 3:
            result.append(lemmatize_stemming(token))
    return result
```

## Create a dictionary out of the given text corpus


```python
import nltk
nltk.download('wordnet')

processed_docs = documents['abstract'].map(preprocess)

print(processed_docs)
```

## Creating a Dictionary from Processed Documents

```python
# Create a dictionary and corpus
dictionary = Dictionary(processed_docs)

print(dictionary)
```

## Using the dictionary, let us convert the documents to bag of words approach

```python
corpus = [dictionary.doc2bow(text) for text in processed_docs]

# let's print the first document in bag of words format
print(corpus[0])

print(len(corpus))
```

# Build the LDA model using the gensim library
we experiment by changing the number of topics in the following code.

Initial Training (6 Topics and 5 topics):

The model was initially trained with 6 and 5 topics to understand the broad distribution of terms and the underlying thematic structure.
Visualization of the results (Intertopic Distance Map) revealed significant overlap between some topics, indicating that these topics were not well-separated.

Revised Training (4 Topics):

To achieve more distinct and interpretable topics, the number of topics was reduced to 4.
The visualization of 4 topics showed minimal overlap between topics, ensuring better differentiation and thematic clarity.

Reason for Adjustment:

Overlap in topics can dilute the interpretability of results, as the same terms may dominate multiple topics.
By reducing the number of topics, the model provides a clearer distinction between themes, which aligns better with the goals of topic modeling.

```python
# Train LDA model
lda_model = LdaModel(corpus=corpus, id2word=dictionary, num_topics=4, passes=10)
```

## Extracting and Assigning Dominant Topics to Documents
This code identifies the dominant topic for each document in the corpus and adds it as a new column to the DataFrame, making it easier to analyze the topic distribution across the dataset.


```python
# Get the topic distribution for each document
topic_distribution = lda_model.get_document_topics(corpus)

# Create a list to store the dominant topic for each document
dominant_topics = []
for distribution in topic_distribution:
    dominant_topic = max(distribution, key=lambda item: item[1])[0]
    dominant_topics.append(dominant_topic+1)

# Add the dominant topic as a new column to the DataFrame
data['dominant_topic'] = dominant_topics

# Now you can access the dominant topic for each abstract using data['dominant_topic']
data.head()
```

## Retrieving Documents Related to a Specific Topic
This code extracts documents associated with a specific topic (e.g., Topic 2) based on a defined probability threshold. It identifies and lists documents that are most relevant to the chosen topic, along with their topic probabilities, enabling focused analysis of topic-specific content.

```python
# Get all abtsracts as list

texts = documents["abstract"].values


# Get documents related to a specific topic (e.g., Topic 0)
topic_id = 2
threshold = 0.5  # Minimum contribution for topic relevance
relevant_docs = []

for i, doc_bow in enumerate(corpus):
    # Get the topic distribution for the document
    doc_topics = lda_model.get_document_topics(doc_bow)

    # Check if the target topic exceeds the threshold
    for topic, prob in doc_topics:
        if topic == topic_id and prob >= threshold:
            relevant_docs.append((i, texts[i], prob))  # Store document index, content, and topic probability
            # relevant_docs.append((i, documents["Abstract"].values[i], prob))  # Store document index, content, and topic probability

# Display relevant documents
for doc_index, content, prob in relevant_docs:
    print(f"Document {doc_index}: {content} (Topic Probability: {prob:.2f}) \n")
```

```python
def get_summary(topic_num):
    filtered_data = data[data["dominant_topic"] == topic_num]
    concatenated_string = ". ".join(filtered_data["abstract"])
    response = model.generate_content(
        "Provide a summary with bullet points for all the articles together:" + concatenated_string,
        generation_config=genai.types.GenerationConfig(
            max_output_tokens=1000,
            temperature=0,
        ),
    )
    return response.text

# Create a list to store the summaries
summaries = []

# Iterate through the unique topics
for topic in data['dominant_topic'].unique():
    summary = get_summary(topic)
    summaries.append(summary)

# Create a dictionary to map topics to summaries
topic_summary_map = dict(zip(data['dominant_topic'].unique(), summaries))

# Add a new column 'summary' to the DataFrame
data['summary'] = data['dominant_topic'].map(topic_summary_map)

data.head()
```
